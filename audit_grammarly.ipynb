```python
def clickPlagiarismButton(second_to_wait):
    # wait 10 seconds for the page to load before getting the source
    sleep(10)
    page_source = BeautifulSoup(driver.page_source)
    
    # this is actually only needed when we use the same doc
    plagiarism_button_off = page_source.find('div',{'class':'fal6plv _4daed495-navigation-outcomeBtn _c6a22528-navigation-outcomeBtnBorders _0485f5a1-document_actions-plagiarismBtn _d4bae220-document_actions-outcomeBtnLeft f1usxtfx', 'data-active':'false'})

    if plagiarism_button_off != None:
        print(f'   5. turning to pla check mode ...')
        plagiarism_button = driver.find_element_by_xpath('/html/body/div/div/div[2]/div[2]/div/div[4]/div[2]/div/div[2]/div/div[2]/div/div[4]/div[2]')
        # plagiarism_button.click()
        driver.execute_script("arguments[0].click();", plagiarism_button)

        # wait 10 sec for grammarly to check the pla
    print(f'   5. checking ({second_to_wait} seconds) ...')
    sleep(second_to_wait)
    print('   6. completed the check')
```
